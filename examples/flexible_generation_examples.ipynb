````python
# å®šç¾©å¤šçµ„ keywords
prompts = [
    {
        "keywords": ["morning", "sunrise", "fresh start"],
        "style": "bright and cheerful"
    },
    {
        "keywords": ["afternoon", "tea time", "relaxing"],
        "style": "warm and cozy"
    },
    {
        "keywords": ["night", "stars", "dreaming"],
        "style": "peaceful and mystical"
    }
]

# æ‰¹æ¬¡ç”Ÿæˆ
results = generator.batch_generate(
    prompts=prompts,
    media_type="image",
    base_config={
        "character": "kirby",
        "system_prompt": "stable_diffusion_prompt",
        "num_images": 2
    }
)

print(f"\næ‰¹æ¬¡ç”Ÿæˆå®Œæˆï¼ç¸½å…±ç”Ÿæˆäº† {sum(len(r['result']['media_files']) for r in results)} å¼µåœ–ç‰‡")
````
